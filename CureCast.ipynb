```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, r2_score
import numpy as np
import matplotlib.pyplot as plt
models={
    "SVC":SVC(kernel="linear"),
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GrandientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNeighbors":KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB":MultinomialNB()
}
for model_name, model in models.items():
    #train model
    model.fit(x_train,y_train)
    #test model
    predictions=model.predict(x_test)
    #calculate accuracy
    accuracy=round(accuracy_score(y_test,predictions),2)
    #confusion matrix
    cm=confusion_matrix(y_test,predictions)
    print(f"{model_name}accuracy:{accuracy}")
    print(f"{model_name}confusion Matrix:")
    print(np.array2string(cm,separator=","))
   
    print("="*30)
    # mae=sklearn.metrics.mean_absolute_error(y_test,predictions)
    # print(f"{model_name}mae:{mae}")
    # mse=sklearn.metrics.mean_squared_error(y_test,predictions)
    # print(f"{model_name}mse:{mse}")
    # rmse=math.sqrt(mse)
    # print(f"{model_name}rmse:{rmse}")
    # rs=r2_score(y_test,predictions)
    # print(f"{model_name}rs:{rs}")
    plt.imshow(cm,interpolation='nearest',cmap='Blues')
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```

```python
symptoms_dict1={key.lower(): value for key, value in symptoms_dict.items()}
print(symptoms_dict1)                                
```

```python
medicine = pd.read_csv('medicine_data1.csv')
diet = pd.read_csv("diet_data.csv")
precautions = pd.read_csv("pre_data.csv")
```

```python
def helper(dis):
     med=medicine[medicine['Disease']==dis] ['Medicine']
     med=[med for med in med.values]
     die=diet[diet['Disease']==dis] ['Diets']
     die=[die for die in die.values]
     pre=precautions[precautions['Disease']==dis] [['Precaution_1','Precaution_2','Precaution_3','Precaution_4']]
     pre=[col for col in pre.values]
     return med,die,pre
```

```python
def patients_values(symptoms):
    symptom_array=np.zeros(len(symptoms_dict1))
    for items in symptoms:
         symptom_array[symptoms_dict1[items]]=1
    return diseases_list[svc.predict([symptom_array])[0]]
```

```python
symp=input("enter your symptoms:")
symptoms=symp.lower()
user_symptoms=[s.strip() for s in symptoms.split(',') ]
user_symptoms=[sym.strip("[]' ") for sym in user_symptoms ]
predicted_diseases=patients_values(user_symptoms)
med,die,pre=helper(predicted_diseases)

print("================================predicted diseases=====================================")
print(predicted_diseases)
print("================================Medicine============================================")
i=1
for m in med:
    print(i,":",m)
    i+=1

print("================================Diet============================================")
i=1
for d in die:
    print(i,":",d)
    i+=1

print("================================precautions============================================")
i=1
for p in pre[0]:
    print(i,":",p)
    i+=1
```
